# Divide Training and Testing Set

## import library

```python
import pandas as pd
import numpy as np
import copy
from collections import Counter
from sklearn.model_selection import train_test_split
```

## divide data

```python
original_data = pd.read_csv("[1]DATA/[1]Original_Data/Hypoxemia.csv")
original_data.info()
```

```python
data = copy.deepcopy(original_data)
data.info()
```

```python
# ratio = 0.2 (testing/whole)
train_data_set, test_data_set = train_test_split(data, test_size=0.2, shuffle=True, stratify=data["Label_Intraoperative"],random_state=0)
train_data_set.info()
print(' ')
print('=========================================== ')
print(' ')
test_data_set.info()
```

# 

```python
# save
train_data_set.to_csv("[1]DATA/[2]Training_and_Testing_Set/training.csv")
test_data_set.to_csv("[1]DATA/[2]Training_and_Testing_Set/testing.csv")
```
