### LASSO，10-折交叉验证寻找最佳参数

```python
"""定义函数"""
def Lasso_CV(x, y, alphas=[], ct=10, cv=5, max_iter=100, normalize=False, fit_intercept=True, path=''):
    
    LASSO_cv = None
    file_name = None
    Number = None
    Path = None
    Best_alpha = []
    Cycle_time = []
    
    for times in range(ct):
        Number = times + 1
    
        LASSO_cv = LassoCV(alphas=alphas, cv=cv, max_iter=max_iter, normalize=normalize, fit_intercept=fit_intercept)
        LASSO_cv.fit(x, y)
    
        file_name = ["LASSO_CV_%s.pkl"%(Number)]
        Path = path + file_name[0]
        joblib.dump(LASSO_cv, Path)
    
        Best_alpha.append(LASSO_cv.alpha_)
        Cycle_time.append(Number)
    
        LASSO_cv = None
        file_name = None
        Number = None
        Path = None
        
    print(pd.Series(index=Cycle_time, data= Best_alpha))
    
    return None   
```

```python
"""术中低氧"""
# 术前变量
# 0.04891
Alphas = np.logspace(-3, 1, 500)
Lasso_CV(x_data_1_standard, y_data_1, alphas=Alphas, ct=20, cv=5, max_iter=500, path='[2]Model_Save/[1]LASSO/[1]Intraoperative/[1]/')
```

```python
# 术前+术中变量
# 0.006541
# 0.005075
Alphas = np.logspace(-3, 1, 500)
Lasso_CV(x_data_2_standard, y_data_2, alphas=Alphas, ct=20, cv=5, max_iter=500, path='[2]Model_Save/[1]LASSO/[1]Intraoperative/[2]/')
```

```python
"""诱导低氧"""
# 术前变量
# 0.005169
Alphas = np.logspace(-3, 1, 500)
Lasso_CV(x_data_3_standard, y_data_3, alphas=Alphas, ct=20, cv=5, max_iter=500, path='[2]Model_Save/[1]LASSO/[2]Induction/[1]/')
```
