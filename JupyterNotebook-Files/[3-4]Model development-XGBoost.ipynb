```python
"""术中低氧——术前变量"""
# 首先通过CV方法确定最佳 n_estimators 的大概范围
xgb_score = []
for i in np.linspace(10,1000,100,endpoint=True):
    j = int(i)
    XGBoost_CV = XGBClassifier(n_estimators=j, 
                               learning_rate=0.1, 
                               max_depth=5, 
                               gamma=0, 
                               min_child_weight=1, 
                               subsample=0.8, 
                               colsample_bytree=0.8, 
                               reg_alpha=0, 
                               reg_lambda=1, 
                               scale_pos_weight=weight_a, 
                               random_state=42)
    score = cross_val_score(XGBoost_CV, x_train_1_standard, y_train_1,cv=10, scoring='roc_auc').mean()
    xgb_score.append(score)
print("最佳AUROC为：%.4f" % max(xgb_score))
print("最佳n_estimators值：",([*range(10,1001,10)][xgb_score.index(max(xgb_score))]))
plt.figure(figsize=[20,5])
plt.plot(range(10,1001,10),xgb_score)
plt.show()
```

```python
# 学习曲线的改进
rs = []
var = []
ge = []

for i in range(39, 60):
    XGBoost_CV = XGBClassifier(n_estimators=i+1, 
                               learning_rate=0.1, 
                               max_depth=5, 
                               gamma=0, 
                               min_child_weight=1, 
                               subsample=0.8, 
                               colsample_bytree=0.8, 
                               reg_alpha=0, 
                               reg_lambda=1, 
                               scale_pos_weight=weight_a, 
                               random_state=42)
    score = cross_val_score(XGBoost_CV, x_train_1_standard, y_train_1, cv=10, scoring='roc_auc')
    #记录1-偏差
    rs.append(score.mean())
    #记录方差
    var.append(score.var())
    #计算泛化误差的可控部分
    ge.append((1 - score.mean())**2+score.var())
#打印R2最高所对应的参数取值，并打印这个参数下的方差
print(range(40, 61)[rs.index(max(rs))],max(rs),var[rs.index(max(rs))])
#打印方差最低时对应的参数取值，并打印这个参数下的R2
print(range(40, 61)[var.index(min(var))],rs[var.index(min(var))],min(var))
#打印泛化误差可控部分的参数取值，并打印这个参数下的R2，方差以及泛化误差的可控部分
print(range(40, 61)[ge.index(min(ge))],rs[ge.index(min(ge))],var[ge.index(min(ge))],min(ge))
plt.figure(figsize=(20,5))
plt.plot(range(40, 61), rs, color="k", label="XGBoost")
# 添加方差线
rs_var_add = [rs[l]+var[l] for l in range(min(len(rs),len(var)))]
rs_var_subtract = [rs[l]-var[l] for l in range(min(len(rs),len(var)))]
plt.plot(range(40, 61), rs_var_add, color="c", linestyle='-.')
plt.plot(range(40, 61), rs_var_subtract, color="m", linestyle='-.')
plt.legend()
plt.show()
```
